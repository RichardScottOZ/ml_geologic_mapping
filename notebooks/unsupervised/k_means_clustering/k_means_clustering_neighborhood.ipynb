```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin
from sklearn.utils import shuffle
from time import time
from skimage import color
from skimage import io

from scipy import ndimage
from scipy import misc
```

## Load useful functions

```python
# %load ../../data_preprocessing/functions.py
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

# A function to plot a matrix
def plot_image(data, colorspace='binary'):
    fig = plt.figure(figsize=(10,10)) 
    ax = fig.add_subplot(111)
    ax.imshow(data, cmap=colorspace)
    
    
def normalize_array(array):
    array_min = array.min(axis=(0, 1), keepdims=True)
    array_max = array.max(axis=(0, 1), keepdims=True)
    normalized_array = (array - array_min)/(array_max - array_min)
    
    return normalized_array

def recreate_image(labels, w, h):
    """Recreate the (compressed) image from the code book & labels"""
    image = np.zeros((w, h))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            image[i][j] = labels[label_idx]
            label_idx += 1
    return image
```

## Read data

```python
image = np.load('../../data_preprocessing/numpy_data/lab_filtered_data.npy')
image.shape
```

## Normalize data for preparation of K-Means

```python
normalized_image = normalize_array(image)
```

```python
w, h = tuple(normalized_image.shape)
```

## Flatten the image, to pass through K-Means algorithm

```python
flattened_image_array = (normalized_image.flatten()).reshape(-1, 1)
print(flattened_image_array.shape)
```

## Execute K-Means classification

```python
n_facies = 10
```

```python
t0 = time()
# Shuffle the array of samples
image_array_sample = shuffle(flattened_image_array, random_state=0)[:150000]
# Fit the K-Means with the subset of the image
kmeans = KMeans(n_facies, random_state=0).fit(image_array_sample)
print("done in %0.3fs." % (time() - t0))
```

```python
t0 = time()
labels = kmeans.predict(flattened_image_array)
print("done in %0.3fs." % (time() - t0))
print(labels.shape)
```

## Recreate the image

```python
t0 = time()
clusters = recreate_image(labels, w, h)
print("done in %0.3fs." % (time() - t0))
```

## Display the results

```python
plt.imshow(image)
plt.show()
plt.imshow(clusters)
plt.show()
```
